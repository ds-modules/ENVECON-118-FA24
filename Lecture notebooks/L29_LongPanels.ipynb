```r
library(haven)
library(tidyverse)
crimedata<- read_dta("CRIME4.DTA")
head(crimedata)
```

```r
head(crimedata[,c("county","lcrmrte", "clcrmrte", "year", "d83", "d84")])
```

```r
#run levels regression
reg0 <- lm(lcrmrte ~ lprbarr + lprbconv + lprbpris + lavgsen + lpolpc + d82 + d83 + d84 + d85 + d86 + d87, data = crimedata)
summary(reg0)
```

```r
#run first differenced regression
reg1<- lm(clcrmrte~ clprbarr + clprbcon + clprbpri + clavgsen + clpolpc + d83 + d84 + d85 + d86 + d87, data = crimedata)
summary(reg1)
```

```r
#run fixed effects regression
crimedata$county<-as.factor(crimedata$county)
reg2<-lm(lcrmrte ~ lprbarr + lprbconv + lprbpris + lavgsen + lpolpc + d82 + d83 + d84 + d85 + d86 + d87 + county , data = crimedata)
summary(reg2)
```

```r
#run fixed effects regression after differencing off means
crimedata<- crimedata %>%
group_by(county) %>%
mutate(meanlcrmrte = mean(lcrmrte, na.rm = T)) %>%
mutate(meanlprbarr = mean(lprbarr, na.rm = T)) %>%
mutate(meanlprbconv = mean(lprbconv, na.rm = T)) %>%
mutate(meanlprbpris = mean(lprbpris, na.rm = T)) %>%
mutate(meanlavgsen = mean(lavgsen, na.rm = T)) %>%
mutate(meanlpolpc = mean(lpolpc, na.rm = T)) %>%
mutate(dlcrmrte = lcrmrte-meanlcrmrte) %>%
mutate(dlprbarr = lprbarr-meanlprbarr) %>%
mutate(dlprbconv = lprbconv-meanlprbconv) %>%
mutate(dlprbpris = lprbpris-meanlprbpris) %>%
mutate(dlavgsen = lavgsen-meanlavgsen) %>%
mutate(dlpolpc = lpolpc-meanlpolpc) %>%
mutate (dw82 = d82-1/7) %>%
mutate (dw83 = d83-1/7) %>%
mutate (dw84 = d84-1/7) %>%
mutate (dw85 = d85-1/7) %>%
mutate (dw86 = d86-1/7) %>%
mutate (dw87 = d87-1/7)


regFEwin <-lm(dlcrmrte~ dlprbarr + dlprbconv + dlprbpris + dlavgsen + dlpolpc + dw82 + dw83 + dw84 + dw85 + dw86 + dw87 + 0, data = crimedata)
summary(regFEwin)

```

```r
#run fixed effects regression using felm
library(lfe)
reg3 <- felm(lcrmrte~ lprbarr + lprbconv + lprbpris + lavgsen + lpolpc + d82 + d83 + d84 + d85 + d86 + d87|county, data = crimedata)
summary(reg3)
```

```r
#compute clustered standard errors using vcovCL - can be used for first differences or fixed effects
library(sandwich)
#this command outputs the variance-covariance matrix.  The diagonal elements of this matrix are the variances
#of the beta coefficients.  If we take the square root of these variances we get our standard errors.
reg2$clus_se <- sqrt(diag(vcovCL(reg2, cluster = crimedata$county)))
#cbind will "bind" the column of coefficients to the column of clustered standard errors
cbind(reg2$coefficients, reg2$clus_se)

```

```r
#compute clustered standard errors in fixed effects regression using felm
reg4 <- felm(lcrmrte~ lprbarr + lprbconv + lprbpris + lavgsen + lpolpc + d83 + d84 + d85 + d86 + d87|county|0|county, data = crimedata)
summary(reg4)
```
